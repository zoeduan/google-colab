#### Simple query

````python
# import openai
from openai import OpenAI
client = OpenAI(api_key=openai_api_key)

def generate_text(prompt, max_tokens=500, temperature=2.0, top_p=1.0, frequency_penalty=0.0, presence_penalty=0.0):
    """
    Generate text based on input prompt using OpenAI's GPT models.

    Args:
        prompt (str): Input text to generate from.
        max_tokens (int): Maximum number of tokens in the generated response.
                         One token is roughly 4 characters or 3/4 of a word.
        temperature (float): Controls randomness (0.0 to 2.0, higher = more random).
        top_p (float): Nucleus sampling parameter (0.0 to 1.0).
                       Limits tokens to the most likely ones that add up to top_p probability mass.
        frequency_penalty (float): Penalizes repeated tokens (-2.0 to 2.0).
        presence_penalty (float): Penalizes repeated topics (-2.0 to 2.0).

    Returns:
        str: Generated text.
    """
    response = client.chat.completions.create(
        model="gpt-4",
        messages=[{"role": "user", "content": prompt}],
        max_tokens=max_tokens,  # This controls the maximum length of the response
        temperature=temperature,
        top_p=top_p,
        frequency_penalty=frequency_penalty,
        presence_penalty=presence_penalty,
        n=1
    )
    return response.choices[0].message.content.strip()
````

````python
prompt = "Artificial intelligence is"
response = generate_text(prompt)
print("Response:", response)
````

### Parameters in LLM

#### 1. Temperature

When a language model generates text, it typically does so by sampling from a probability distribution over possible next tokens. Temperature is a hyperparameter that modifies how much randomness‚Äîor ‚Äúcreativity‚Äù‚Äîthe model exhibits when sampling from this distribution.

Mathematically, temperature T rescales the logits (i.e., raw output scores) before they are converted into probabilities via the softmax function.

- T=1: The logits remain unchanged, and the original probabilities are used for sampling.
- High Temperature (T>1): The scaling effectively ‚Äúflattens‚Äù the distribution, making less probable tokens more likely to be sampled. This results in more ‚Äúrandom‚Äù or ‚Äúcreative‚Äù outputs.
- Low Temperature (0<T<1): The distribution becomes ‚Äúsharper,‚Äù concentrating the probability mass on the most likely tokens. This results in more ‚Äúdeterministic‚Äù outputs, where the language model is more likely to pick the most probable word according to its training data.

#### 2. Top-p

An alternative to sampling with temperature, called nucleus sampling, where the model considers the results of the tokens with top-p probability mass. So 0.1 means only the tokens comprising the top 10% probability mass are considered.


It sorts tokens by probability from highest to lowest. Selects the smallest set of tokens that together reach a specified cumulative probability p. Renormalizes within that set and samples.



#### Putting Top-P and Temperature Together

When You Want Simple, Global Control ‚Üí Use Temperature

* Temperature changes the shape of the entire probability distribution.
* Easy to dial up (creative) or dial down (deterministic).

<!-- When You Want a Fixed Pool of Candidates ‚Üí Use Top-k
* You know exactly how many tokens will be considered each time.
* Good if you need guaranteed ‚Äúdepth‚Äù of exploration up to kk tokens, but no further. -->

When You Want Probability-Based Adaptation ‚Üí Use Top-p
* The number of tokens can shrink or expand depending on how certain the model is at each step.
* Often a sweet spot for balancing variety with coherence because it adapts automatically to the model‚Äôs underlying confidence.

#### 3. Frequency_penalty
Reduces repetition of the same tokens. Number between -2.0 and 2.0. Positive values penalize new tokens based on their existing frequency in the text so far, decreasing the model's likelihood to repeat the same line verbatim.

#### 4. Presence_penalty
Reduces repetition of similar topics. Number between -2.0 and 2.0. Positive values penalize new tokens based on whether they appear in the text so far, increasing the model's likelihood to talk about new topics.

#### 5. max_tokens
The maximum number of tokens that can be generated in the chat completion. This value can be used to control costs for text generated via API. Please note, max_tokens is not compatible with o1 series models.

### Play with the parameters in ChatGPT


````python
from ipywidgets import interact, FloatSlider, IntSlider, Text, Dropdown
@interact(
    model=Dropdown(
        options=['gpt-4', 'gpt-4-turbo', 'gpt-3.5-turbo'],
        value='gpt-4',
        description='Model:'
    ),
    temperature=FloatSlider(min=0.0, max=2.0, value=0.7, step=0.1, description='Temperature:'),
    max_length=IntSlider(min=20, max=500, value=100, description='Max Length:'),
    top_p=FloatSlider(min=0.1, max=1.0, value=1.0, step=0.1, description='Top-p:'),
    frequency_penalty=FloatSlider(min=-2.0, max=2.0, value=0.0, step=0.1, description='Freq Penalty:'),
    presence_penalty=FloatSlider(min=-2.0, max=2.0, value=0.0, step=0.1, description='Pres Penalty:'),
)
def interactive_generate(
    prompt="The meaning of life is",
    model="gpt-4",
    temperature=0.7,
    max_tokens=500,
    top_p=1.0,
    frequency_penalty=0.0,
    presence_penalty=0.0,
):
    """
    Interactive text generation using OpenAI's GPT models.

    Args:
        prompt (str): Input text to generate from
        model (str): OpenAI model to use
        temperature (float): Controls randomness
        max_tokens (int): Maximum token of response
        top_p (float): Nucleus sampling parameter
        frequency_penalty (float): Penalizes token repetition
        presence_penalty (float): Penalizes topic repetition
    """
    try:
        # Prepare messages
        messages = [{"role": "user", "content": prompt}]

        # Call OpenAI API
        response = client.chat.completions.create(
            model=model,
            messages=messages,
            max_tokens=max_tokens,
            temperature=temperature,
            top_p=top_p,
            frequency_penalty=frequency_penalty,
            presence_penalty=presence_penalty,
            n=1,
            stream=False,
        )

        # Print the generated text
        print(f"\n--- Generated Text ---\n")
        print(response.choices[0].message.content)
        # print(f"\n--- End of Text ---\n")

    except Exception as e:
        print(f"Error: {e}")

# For standalone prompt entry (optional)
prompt_widget = Text(
    value='The meaning of life is',
    placeholder='Enter your prompt here',
    description='Prompt:',
    disabled=False,
    style={'description_width': 'initial'},
    layout={'width': '90%'}
)

def on_submit(sender):
    interactive_generate(prompt=prompt_widget.value)

prompt_widget.on_submit(on_submit)
````

````python
import requests
import json

url = "http://localhost:11434/api/chat"
````

````python
def llama3(prompt):
    data = {
        "model": "llama3-parameters",
        "messages": [
            {
                "role": "user",
                "content": prompt
            }
        ],
        "stream": False,
    }

    headers = {
        "Content-Type": "application/json"
    }

    response = requests.post(url, headers=headers, json=data)
    response_json = response.json()

    # Print the full response to understand the structure
    print("Full response:", response_json)

    # # Handle the response structure correctly
    if "response" in response_json:
        return response_json["response"]
    elif "message" in response_json and "content" in response_json["message"]:
        return response_json["message"]["content"]
    else:
        return "Unable to parse response: " + str(response_json)
````

# 3. Incorporating generative AI into research

## 3.1 General best practices to follow

### Verify and Fact-Check AI Outputs

- AI-generated content, while persuasive, may contain inaccuracies, incomplete details, or fabricated information (‚Äúhallucinations‚Äù).

- Cross-reference AI-generated data, summaries, and analytical results with peer-reviewed literature, OSU libraries' resources, and trusted primary sources.

- Always manually verify AI-generated citations and references through OSU's library databases, as AI tools may generate incorrect or fictional citations.

### Maintain Confidentiality and Data Security

- Do not input sensitive, unpublished, proprietary, or restricted OSU information into AI tools.

- Utilize OSU-approved, locally hosted, or secure AI models with enhanced privacy protections when handling confidential data.

- For research involving human subjects, confirm AI tool compliance with OSU Institutional Review Board (IRB) protocols and data privacy regulations such as GDPR and HIPAA.

### Understand AI's Limitations with Language and Context

- Be cautious with AI-generated translations, as technical accuracy, cultural nuances, and contextual tone may be compromised.

- In specialized research areas, seek human expert verification through OSU's network of domain specialists to ensure accurate translation and context.

- Exercise additional caution when generating content in languages beyond the AI's primary training languages.

### Clearly Attribute and Document AI Use

- Transparently acknowledge AI-generated contributions in OSU research publications, presentations, and other relevant outputs.

- Adhere to OSU and publisher guidelines regarding the citation and attribution of AI-generated materials.

- Clearly document and describe the extent and nature of AI assistance within the research methodology (e.g., data analysis support, drafting assistance, hypothesis development).

### Use AI as a Tool‚ÄîNot a Replacement for Critical Thinking

- Leverage AI for automating routine research tasks such as data summarization and coding, while relying on your critical judgment for research analysis and interpretation.

- Critically evaluate AI-generated conclusions and validate them against traditional methodologies and expert consultation within OSU's academic community.

- Integrate AI-generated insights thoughtfully alongside established research practices to achieve rigorous and reliable outcomes.

### Be Mindful of Overreliance on AI

- Utilize AI to enhance efficiency but do not substitute it for comprehensive domain knowledge and intellectual evaluation.

- Create balanced workflows that include rigorous human oversight to ensure ongoing research quality and integrity.

- Regularly evaluate AI's impact on your research processes and adjust usage to maintain OSU's standards for scholarly excellence.

### Stay Current with AI Developments and OSU Institutional Guidelines

- AI technologies evolve quickly; regularly update your understanding of the capabilities and limitations of the AI tools you utilize.

- Stay informed about updates in training data, algorithmic improvements, and privacy policy changes relevant to AI tools.

- Frequently consult OSU's policies and guidelines on AI usage in research to ensure full alignment with the university's academic integrity and research ethics standards.

````python
from google.colab import userdata

GOOGLE_API_KEY=userdata.get('GOOGLE_API_KEY')
````

````python
from google import genai
from google.genai import types

client = genai.Client(api_key=GOOGLE_API_KEY)
````

````python

MODEL_ID = "gemini-1.5-flash-latest" # @param ["gemini-1.5-flash-latest","gemini-2.0-flash-lite","gemini-2.0-flash","gemini-2.0-pro-exp-02-05"] {"allow-input":true, isTemplate: true}
````

üí° Please try changing the MODEL_ID

````python
def askGemini(content, model_id=MODEL_ID):
    response = client.models.generate_content(
    model=MODEL_ID,
    contents="Tell me how the internet works, but pretend I'm a puppy who only understands squeaky toys.",
    config=types.GenerateContentConfig(
        temperature=0.4,
        top_p=0.95,
        top_k=20, # Limits token selection to the top 20 most probable tokens
        candidate_count=1, # Specifies how many response candidates the model should generate. A value of 1 means only one response will be returned
        seed=5, # Sets a fixed random seed for reproducibility. Using the same seed with the same prompt ensures the same output.
        max_output_tokens=100,
        stop_sequences=["STOP!"], # Defines a stop condition. If the model generates "STOP!", it will terminate the response immediately.
        presence_penalty=0.0,
        frequency_penalty=0.0,
    )
)
    return response.text
````

````python
print(askGemini("Tell me how the internet works, but pretend I'm a puppy who only understands squeaky toys."))
````

### Gemma

Gemma is a family of lightweight, state-of-the-art open models from Google, built from the same research and technology used to create the Gemini models. Gemma 3 models are multimodal, handling text and image input and generating text output, with open weights for both pre-trained variants and instruction-tuned variants. Gemma 3 has a large, 128K context window, multilingual support in over 140 languages, and is available in more sizes than previous versions. Gemma 3 models are well-suited for a variety of text generation and image understanding tasks, including question answering, summarization, and reasoning. Their relatively small size makes it possible to deploy them in environments with limited resources such as laptops, desktops or your own cloud infrastructure, democratizing access to state of the art AI models and helping foster innovation for everyone.

````python
import requests
import json

url = "http://localhost:11434/api/chat"
````

````python
def gemma3(prompt):
    data = {
        "model": "gemma3:4b",
        "messages": [
            {
                "role": "user",
                "content": prompt

            }
        ],
        "stream": False,
    }

    headers = {
        "Content-Type": "application/json"
    }

    response = requests.post(url, headers=headers, json=data)
    return response.json()["message"]["content"]
````

````python
response = gemma3("Can you give a brief explaination of the paper 'Attention is all your need'?")
print(response)
````
